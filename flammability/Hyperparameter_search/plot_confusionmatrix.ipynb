```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, auc

# 读取 CSV 文件

df_conf_test = pd.read_csv('./XGBoost_overall_conf_matrix_test.csv', index_col=0)

```

```python
, 'fontweight': 'bold'
```

```python
# 绘制混淆矩阵
conf_matrix_test = df_conf_test.values  # 获取混淆矩阵数据

plt.figure(figsize=(6, 5),dpi=500)
sns.heatmap(conf_matrix_test, annot=True, fmt='d',annot_kws={'fontsize': 30}, cmap='Blues', xticklabels=['Predicted 0', 'Predicted 1'], yticklabels=['True 0', 'True 1'])
plt.title('XGBoost',fontsize=20)
plt.xlabel('Predicted',fontsize=25)
plt.ylabel('True',fontsize=25)
plt.xticks(fontsize=20)  # 调整X轴标签文字大小
plt.yticks(fontsize=20)  # 调整Y轴标签文字大小
plt.show()
```

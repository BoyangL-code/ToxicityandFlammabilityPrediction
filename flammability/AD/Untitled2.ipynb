```python
import numpy as np
import matplotlib.pyplot as plt

#  (Low Threshold)   (High Threshold)
low_values = np.array([0.20, 0.25, 0.30, 0.35, 0.40])
high_values = np.array([0.60, 0.65, 0.70, 0.75, 0.80])

#  (Accuracy Matrix)
accuracy_matrix = np.array([
    [0.970, 0.970, 0.970, 0.970, 0.444],
    [0.970, 0.970, 0.970, 0.970, 0.667],
    [0.969, 0.969, 0.969, 0.969, 'nan'],
    [0.967, 0.967, 0.967, 0.967, 'nan'],
    [0.966, 0.966, 0.966, 0.966, 'nan']
])

# 
fig, ax = plt.subplots(figsize=(8, 6), dpi=500)

# 
cax = ax.imshow(accuracy_matrix, cmap='Blues', aspect='auto', origin='upper')

# 
cbar = plt.colorbar(cax, label='Accuracy')

# 
ax.set_xticks(np.arange(len(high_values)))
ax.set_yticks(np.arange(len(low_values)))
ax.set_xticklabels([f"{v:.2f}" for v in high_values], fontsize=12)
ax.set_yticklabels([f"{v:.2f}" for v in low_values], fontsize=12)

# 
ax.set_xlabel("High Threshold", fontsize=14)
ax.set_ylabel("Low Threshold", fontsize=14)

# 
for i in range(len(low_values)):
    for j in range(len(high_values)):
        value = accuracy_matrix[i, j]
        color = 'white' if value > np.mean(accuracy_matrix) else 'black'
        ax.text(j, i, f"{value:.3f}", ha='center', va='center', color=color, fontsize=12)

# 
plt.show()
```

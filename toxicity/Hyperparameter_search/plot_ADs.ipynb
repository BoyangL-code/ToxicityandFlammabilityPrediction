```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 从数据中提取模型名称和评估指标
df = pd.read_excel('./toxicity_evaluation_test_file.xlsx')

models = df['Unnamed: 0']
metrics = ['Toxicity', 'Flammability']

# 提取每个模型的指标值
acc = df['Toxicity']
recall = df['Flammability']

from matplotlib import font_manager

# 创建加粗字体对象
bold_font = font_manager.FontProperties(weight='bold', size=20)
# 设置条形图的宽度和位置
bar_width = 0.2
index = np.arange(len(models))

# 创建图形
fig, ax = plt.subplots(figsize=(12, 6), dpi=500)

# 绘制每个指标的条形图
ax.bar(index - bar_width / 2, acc, bar_width, label='Toxicity', color='#e8eaf6')
ax.bar(index + bar_width / 2, recall, bar_width, label='Flammability', color='#bbcfe8')

# 设置图形的标签
ax.set_xlabel('Models', fontsize=20, fontweight='bold')
ax.set_ylabel('Values', fontsize=20, fontweight='bold')

# 调整 x 轴刻度，使标签居中
ax.set_xticks(index)
ax.set_xticklabels(models, rotation=0, fontsize=20, fontweight='bold')
plt.yticks(fontsize=20, fontweight='bold')
# 设置图例
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.02), ncol=6, frameon=False, prop=bold_font)

# 显示图形
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 从数据中提取模型名称和评估指标
df = pd.read_excel('./Flam_evaluation_test_file.xlsx')

models = df['Unnamed: 0']
metrics = ['ACC', 'REC', 'PREC', 'F1', 'PR-AUC', 'AU-ROC']

# 提取每个模型的指标值
acc = df['ACC']
recall = df['REC']
precision = df['PREC']
f1 = df['F1']
pr_auc = df['PR-AUC']
balanced_acc = df['AU-ROC']  # 假设平衡准确度的列为 'AU-ROC'

# 设置条形图的宽度和位置
bar_width = 0.12
index = np.arange(len(models))

# 创建图形
fig, ax = plt.subplots(figsize=(12, 6),dpi=500)

# 绘制每个指标的条形图
ax.bar(index, acc, bar_width, label='ACC', color='#e8eaf6')
ax.bar(index + bar_width, recall, bar_width, label='REC', color='#bbcfe8')
ax.bar(index + 2 * bar_width, precision, bar_width, label='PREC', color='#92bbd7')
ax.bar(index + 3 * bar_width, f1, bar_width, label='F1', color='#659bca')
ax.bar(index + 4 * bar_width, pr_auc, bar_width, label='PR-AUC', color='#376eae')
ax.bar(index + 5 * bar_width, balanced_acc, bar_width, label='AU-ROC', color='#234086')

# 设置图形的标签
ax.set_xlabel('Models', fontsize=15)
ax.set_ylabel('Values', fontsize=15)
#ax.set_title('Model Evaluation Metrics', fontsize=16, fontweight='bold',fontproperties=prop)
ax.set_xticks(index + 2 * bar_width)  # 将 x 轴标签放置在条形图中间
ax.set_xticklabels(models, rotation=0,fontsize=15)

# 添加图例
#ax.legend(frameon=False)
#, prop=prop
# 设置图例位置在图形最上方，横向排列
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.02), ncol=6, frameon=False, fontsize=12)



# 显示图形
plt.tight_layout()
plt.show()
```
